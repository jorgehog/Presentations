```python
def LJ12_6(r, sigma=1, eps=1):
    
    sr6 = (sigma/r)**6
    
    return 4*eps*sr6*(sr6 - 1)

r = linspace(0.5, 2.5, 200)

text = "$V_\mathrm{LJ}(r_{ij}\,;\,\sigma,\,\epsilon)$"
plot(r, LJ12_6(r), c=myGreen, linewidth=4, label=text)
locs, labels = xticks()

rm =  2**(1./6)
plot([0, 3], [0, 0], 'k--')
plot([1, 1], [-1.5, 0], 'k--')
plot([0, rm], [-1, -1], 'k--')
plot([rm, rm], [-2, -1], 'k--')

xticks([1, rm,  2], [r"$\sigma$", r"$\sigma_m$", r"$2\sigma$"])
yticks([-1, 0, 1], [r"-$\epsilon$", "0", r"$\epsilon$"])

ylim([-1.5, 1.5])
xlim(0.75, 2.5)

xlabel("$r_{ij}$")

legend()

gca().spines['top'].set_visible(False)
gca().spines['right'].set_visible(False)
gca().get_xaxis().tick_bottom()
gca().get_yaxis().tick_left()

show()
    
```
